```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Figure, Layout
import csv

init_notebook_mode(connected=True)

column1, column2 = [], []
column3=[]
global column3

with open('data/cleaned/consolidated.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        column1.append(row['Day of Week'])
        column2.append(row['Registered Pct'])
        column3.append(float(row['Casual'])/float(row['Registered']))

iplot([Scatter(x=column1, y=column2, mode='markers')])
```
